# GETTING LOAN ON BASIS OF CREDIT_SCORE AND PREVIOUS_LOAN TAKEN FROM BANK

**IMPORTIND LIBRARIES , DATASET , DIFFRENTIATING THE DEPENDENT AND INDEPENDENT VARIABLE**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("/content/drive/MyDrive/PREPINSTA /MACHINE LEARNING /4. CLASSIFICATION /DATASET /Loan Distribution KNN Classific.csv")
data_set
```

```python
X = data_set.iloc[:,:-1].values
y = data_set.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

**SPLITING DATA INTO TESTING AND TRAINING DATA**

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.3 , random_state = 42)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

**APPLYING FEATURE SCALING**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(X_train)
x_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

**KNN  TRAINING ON TRAINING SET**

```python
from sklearn.neighbors import KNeighborsClassifier
# we will specify how much naighbours we want to choose
# metric = minkoskki ( WE WILL USE THIS METHOD HERE ) = USED FOR CALCULATING THE EUCLIDEAN DISTANCE
# P =2 IT IS POWER PARAMETER FOR THE MONKASKI METHOD

classifier = KNeighborsClassifier(n_neighbors = 5 , metric = "minkowski" , p = 2 )
classifier.fit(X_train , y_train)

# now our model is ready
```

**predicting getting the loan for the new person based on befor loans and credit score**

```python
print(classifier.predict(sc.transform([[3,730 ]])))
# sc_transform we have used feature scaling above
# so rather than writting .inverse() , .reshape() we have used this
```

**PREDICTING THE Y_PREDICTION VALUES AND Y_TESTING VALUES (TEST_SET RESULT)**

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1 )), 1))
```

**CHEAKING THE CONFUSION MATRIX**

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)


# here 40 times given loan and 28 times we predicted wrongly
```

**VISUALIZING TRAINING SET RESULT**

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 0, stop = X_set[:, 0].max() + 3, step = 1),
                     np.arange(start = X_set[:, 1].min() - 0, stop = X_set[:, 1].max() + 200, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('KNN (Training set)')
plt.xlabel('No of Existing Loans')
plt.ylabel('Credit Score')
plt.legend()
plt.show()     # RED = NOT PROVIDE LOANS
                # GREEN = PROVIDE LOANS
```

**VISUALIZING TEST SET RESULT**

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 0, stop = X_set[:, 0].max() + 3, step = 1),
                     np.arange(start = X_set[:, 1].min() - 0, stop = X_set[:, 1].max() + 200, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('KNN (Test set)')
plt.xlabel('No of Existing Loans')
plt.ylabel('Credit Score')
plt.legend()
plt.show() # RED = NOT PROVIDE LOANS
          # GREEN = PROVIDE LOANS
```
